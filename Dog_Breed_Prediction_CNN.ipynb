### Begining the CNN Model

````python
# Importing the libraries
%%time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm
from keras.preprocessing import image
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
from tensorflow.keras.optimizers import Adam
````

````python
!cd /content/dog_dataset
!ls -lrt
````

````python
# We will read the labels.csv

labels_all = pd.read_csv('dog_dataset/labels.csv')
labels_all['breed'].value_counts()
````

````python
labels_all['breed'].unique()
````

````python
CLASS_NAMES = ['labrador_retriever','doberman','german_shepherd' ]
````

````python
labels = labels_all[labels_all['breed'].isin(CLASS_NAMES)]
````

````python
labels['breed'].value_counts()
````

````python
labels = labels.reset_index()
````

````python
##We will do onehot encoding of the classes
````

````python
len(labels) ##227, we have 227 labels altogether
````

````python
##Creating the numpy matrix with zero
## 1st : Number of labels
## 2nd : Number of pixels in y axis
## 3rd : Number of pixels in x axis
## 4th : For each pixel, these are the number of channels

X_data = np.zeros((len(labels), 224, 224, 3), dtype='float32')
````

````python
X_data.shape
````

````python
# three channels for the pixel at the origin
X_data[0][0][0]
````

````python
#One hot encoding of the labels
y_data = label_binarize(labels['breed'], classes=CLASS_NAMES )
````

````python
## Reading and converting the data into numpy array and normalizing it
````

````python
%%time
from numpy.core.fromnumeric import size
for i in tqdm(range(len(labels))):
  #How to load the image using image.load_img() method in Keras
  img = image.load_img('dog_dataset/train/%s.jpg' %labels['id'][i], target_size= (224,224))
  #print(img.size) #--> We have different sizes of images
  img=image.img_to_array(img) #--> This method convert the image to array with (224,224,3) shape
  #print(img)
  x = np.expand_dims(img.copy(), axis=0)
  X_data[i] = x/255.0
````

````python
tqdm(X_data[0])
````

````python
img
````

````python
## Building the model
````

````python
model = Sequential()
````

````python
%%time
#Step 1 : Layer 1 : Input layer
model.add(Conv2D(filters=64, kernel_size=(5,5), activation='relu', input_shape=(224,224,3)))
model.add(MaxPool2D(pool_size=(2,2)))

#Step 2.1 : Layer 2.1 : CNN 1
model.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu', kernel_regularizer='l2'))
model.add(MaxPool2D(pool_size=(2,2)))

#Step 2.2 : Layer 2.2 : CNN 2
model.add(Conv2D(filters=16, kernel_size=(7,7), activation='relu', kernel_regularizer='l2'))
model.add(MaxPool2D(pool_size=(2,2)))

#Step 2.3 : Layer 2.3 : CNN 3
model.add(Conv2D(filters=8, kernel_size=(5,5), activation='relu', kernel_regularizer='l2'))
model.add(MaxPool2D(pool_size=(2,2)))

#Step 3 : Flattening layer
model.add(Flatten())

#Step 4.1 : MLP Layer
model.add(Dense(128,activation='relu',kernel_regularizer='l2'))
#Step 4.2
model.add(Dense(64,activation='relu',kernel_regularizer='l2'))

#Step 5  : Classification Layer
model.add(Dense(len(CLASS_NAMES),activation='softmax'))

#Step 6 : Colpilation : Mention the Optimizer, loss, metrics etc.
model.compile(optimizer=Adam(0.001), loss='categorical_crossentropy', metrics =['accuracy'])

#Step 7 : Summarize the execution.
model.summary()

````

````python
## Train test split the data

X_train_and_val, X_test, y_train_and_val, y_test = train_test_split(X_data, y_data, test_size=0.1)
````

````python
X_train, X_val, y_train, y_val = train_test_split(X_train_and_val, y_train_and_val, test_size=0.2)
````

````python
%%time
epochs = 100
batch_size = 128

history = model.fit(x=X_train,y=y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_val,y_val))
````

````python
import joblib
````

````python
joblib.dump(history, 'history.pkl')
````

````python
df_model = pd.DataFrame(history.history , columns=history.history.keys())
df_epoch = pd.DataFrame(np.arange(1,len(df_model)+1), columns=['epoch'])
df_model_performance = pd.concat([df_model, df_epoch], axis = 1)
df_model_performance
````

````python
sns.set_style('darkgrid')
fig = plt.figure(figsize=(20,6))
fig.tight_layout(w_pad=5)
ax1 = fig.add_subplot(1,2,1)
sns.lineplot(data = df_model_performance, x ='epoch', y='accuracy', color = '#222f5b', label = 'accuracy' )
sns.lineplot(data = df_model_performance, x ='epoch', y='val_accuracy', color = '#946b2d', label = 'val_accuracy' ) ##loss
#sns.lineplot(data = df_model_performance, x ='epoch', y='loss', color = '#946b2d', label = 'loss' )
plt.title('Model Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel('Epoch',fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)

ax2 = fig.add_subplot(1,2,2)
sns.lineplot(data = df_model_performance, x ='epoch', y='loss', color = '#222f5b', label = 'loss' )
sns.lineplot(data = df_model_performance, x ='epoch', y='val_loss', color = '#946b2d', label = 'val_loss' ) ##loss
#sns.lineplot(data = df_model_performance, x ='epoch', y='loss', color = '#946b2d', label = 'loss' )
plt.title('Loss', fontsize = 14)
plt.legend(fontsize = 14)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel('Epoch',fontsize = 14)
plt.ylabel('Loss',fontsize = 14)
````
